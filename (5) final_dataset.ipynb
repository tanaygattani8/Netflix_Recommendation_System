```python
df3 = pd.read_csv('datasets\American_Movies_2019_to_2025_Master.csv')
```

```python
df3.columns
```

```python
title_duplicates = df3.duplicated(subset=['movie_title']).sum()
print(f"I found {title_duplicates} duplicate movie titles!")
```

```python
df3.duplicated().sum()
```

```python
df3.drop_duplicates(subset=['movie_title'], keep='first', inplace=True)
df3.reset_index(drop=True, inplace=True)
```

```python
df3.isna().sum()
```

```python
df3 = df3.dropna(how='any')
```

```python
df3.isna().sum()
```

```python
df3 = df3.rename(columns={'tmdb_score': 'imdb_score'})
df3['movie_title'] = df3['movie_title'].str.lower()
```

```python
new_df3 = df3.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title','language','duration','imdb_score']]
```

```python
new_df3
```

```python
new_df3['genres'] = new_df3['genres'].str.replace(', ', ' ')
new_df3.head()
```

```python
new_df3['comb'] = new_df3['actor_1_name'] + ' ' + new_df3['actor_2_name'] + ' '+ new_df3['actor_3_name'] + ' '+ new_df3['director_name'] +' ' + new_df3['genres']
new_df3
```

```python
final_data = pd.concat([df1, new_df2, new_df3], ignore_index=True)
final_data
```

```python
final_data.isna().sum()
```

```python
final_data = final_data.dropna(how='any')
```

```python
final_data.isna().sum()
```

```python
final_data.to_csv('datasets/final_data.csv', index=False)
```
